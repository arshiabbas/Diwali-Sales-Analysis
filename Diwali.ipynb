```python
#drop null values
diwali.dropna(inplace=True)
```

```python
#changing the amount data types
diwali['Amount'] = diwali['Amount'].astype('int')
```

```python
#Exploratory Data Analysis
plt.figure(figsize=(4,4))
ax = sns.countplot(data=diwali,x='Gender', hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

```python
sales_by_gender = diwali.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount', data=sales_by_gender, hue='Gender')
plt.show()
```

**From The above two graphs we can see that more customers are Female, and females have purchased more compared to men's**

```python
ax = sns.countplot(x='Age Group',data=diwali, hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

**From Here we can see the more customers are from age 26-35**

```python
#total amount vs age group
sales_by_age = diwali.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount', data=sales_by_age, hue='Age Group')
plt.show()
```

**From the graph we can see most customers who purchased are from agre group 26-35**

**STATE**

```python
#state wise orders
state_orders = diwali.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(8)
plt.figure(figsize=(15,5))
sns.barplot(x='State',y='Orders', data=state_orders, hue='State')
plt.xticks(rotation=0)
plt.show()
```

**Most Customers are from UP, Maharastra and karnataka**

```python
#total amount sales from top states
state_amount = diwali.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(8)
plt.figure(figsize=(15,5))
sns.barplot(x='State',y='Amount', data=state_amount, hue='State')
plt.xticks(rotation=0)
plt.show()
```

```python
#marital Status
ax = sns.countplot(data = diwali, x='Marital_Status', hue='Marital_Status')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_ms = diwali.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount', data=sales_ms, hue='Gender')
plt.show()
```

**From the above graph, most of the buyers are married and are female**

```python
#Occupation
plt.figure(figsize=(15,5))
ax = sns.countplot(data = diwali, x='Occupation', hue='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
plt.xticks(rotation=40)
plt.show()
```

```python
occupation = diwali.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.barplot(x='Occupation',y='Amount', data=occupation, hue='Occupation')
plt.show()
```

**Most of the customers are from IT, Healthcare and Aviation Occupations**

```python
#product category
plt.figure(figsize=(15,4))
ax = sns.countplot(x='Product_Category', data=diwali, hue='Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
plt.xticks(rotation=60)
plt.show()
```

```python
product_car = diwali.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
plt.figure(figsize=(15,4))
sns.barplot(data=product_car, x='Product_Category', y='Amount', hue='Product_Category')
plt.xticks(rotation=70)
plt.show()
```

**Most of the sold procts are Food, Clothing anf Electronics**
